```python
import numpy as np
import matplotlib.pyplot as plt

# Cargar los datos del archivo como strings
data_str = np.loadtxt('C:\\Users\\hugo\\Documents\\lab 3 ondas\\espectroscopia T1.txt', delimiter='\t', dtype=str)

# Reemplazar comas por puntos decimales y convertir a números de punto flotante
data = np.array([[float(s.replace(',', '.')) for s in row] for row in data_str])

# Extraer los ángulos de difracción y las intensidades
angulos = data[:, 0]
intensidades = data[:, 1]

# Graficar los datos
fig, ax = plt.subplots()
ax.bar(angulos, intensidades)
ax.set_xlabel('Ángulo de difracción (grados)')
ax.set_ylabel('Intensidad (unidades arbitrarias)')
ax.set_title('Patrón de interferencia')

plt.show()
```
